```python
import itertools

def implies(x, y):
    return not x or y

def neg(x):
    return not x

def or_op(x, y):
    return x or y

def KB(Q, P, R):
    KB_1 = implies(Q, P)
    KB_2 = implies(P, neg(Q))
    KB_3 = or_op(Q, R)

    return KB_1 and KB_2 and KB_3

def alpha(R):
    # Alpha is R, so we just check if R is True
    return R

# Generate all combinations of truth values for Q, P, R
def create_truth_table():
    return list(itertools.product([False, True], repeat=3))

# Check if KB entails alpha and print the truth table
def check_entailment():
    truth_table = create_truth_table()

    # Print the header for the truth table
    print(f"{'Q':<5}     {'P':<5}     {'R':<5}     {'KB(Q->P, P->~Q, Q U R)':<25}     {'Alpha(R -> P)':<5}")
    print("-" * 75)

    for Q, P, R in truth_table:
        # Evaluate the knowledge base
        KB_is_true = KB(Q, P, R)

        # Evaluate alpha (R)
        alpha_is_true = alpha(implies(R,P))

        # Print the row for the truth table
        print(f"{str(Q):<5}     {str(P):<5}     {str(R):<5}     {str(KB_is_true):<25}     {str(alpha_is_true):<5}")

    # Finally, return whether the KB entails alpha
    for Q, P, R in truth_table:
        # Evaluate the knowledge base
        KB_is_true = KB(Q, P, R)

        # If the KB is true, check if alpha (R) is true
        if KB_is_true and not alpha(implies(R,P)):
            return "No"

    return "Yes"

# Run the entailment check
result = check_entailment()
if result=="Yes":
  print("\nKB entail alpha")
else:
  print("\nKB doesnot entail alpha")
```
